```python
# What is a training a model mean?
# Training a model means teaching it how to make predictions by showing it data. The model looks at input-output pairs and learns patterns or relationships.
# Training = showing the model examples so it can learn.

# What is a fitting a model mean?
# Fitting a model means using training data to help the model learn the relationship between input and output.
# In machine learning, when we say: "Fit a model to data", Find the best parameters (like slope and intercept for a line) that make the model predict the output (y) from the input (X) as accurately as possible.

# You know input and you know output , now teach a model how to get the desired output if given input is supplied.
```

```python
# Linear regression
from sklearn import linear_model
# Creates an instance of the LinearRegression model named reg. reg.fit(X,y)
reg = linear_model.LinearRegression()
reg.fit([[0, 0], [1, 1], [2, 2]], [0, 1, 2])
reg.coef_
reg.intercept_
```
